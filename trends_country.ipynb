## Python code (use in Google Colab) to generate the Global Trends Historically

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (replace 'your_file.csv' with your actual file path)
file_path = '/content/results_country_yearly.csv'
data = pd.read_csv(file_path, header=None)

# Rename columns for clarity
data.columns = ['Year', 'Average_Global_Temp']

# Plot the data
plt.figure(figsize=(15, 5))
plt.plot(data['Year'], data['Average_Global_Temp'], marker='o', linestyle='-', color='b')
plt.title('Average Global Temperature Over Years')
plt.xlabel('Year')
plt.ylabel('Average Global Temperature (°C)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
