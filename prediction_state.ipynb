## Python code (use in Google Colab) to generate the heatmap

Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

Specify the filename (use the name of the uploaded file)

```python
file_name = '/content/results_state.csv'
```

Define the column names

```python
columns = ['state', 'Avg_Temp_Next_Decade']
```

Read the CSV file without headers

```python
data = pd.read_csv(file_name, header=None, names=columns)
```

Transform the data for the heatmap

```python
heatmap_data = data.melt(id_vars=['state'],
                         var_name='Year', value_name='Temperature')
```

Create a pivot table

```python
pivot_table = heatmap_data.pivot(index='state', columns='Year', values='Temperature')
```

Plot the heatmap

```python
plt.figure(figsize=(4, 48))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".1f", linewidths=0.5)
plt.title("Average Temperature Heatmap for Next Decade (2025-2035)")
plt.show()
```
